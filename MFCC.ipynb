```python
import matplotlib.pyplot as plt
import librosa.display
import librosa

def gerarGraficoMFCC(sound_file):
    y, sr = librosa.load(sound_file, 44100)
    mfccs = librosa.feature.mfcc(y = y, sr = sr, n_mfcc=40)

    fig, ax = plt.subplots()
    img = librosa.display.specshow(mfccs, x_axis='tempo', ax=ax)
    fig.colorbar(img, ax=ax)
    ax.set(title='MFCC')
    
```

```python
gerarGraficoMFCC('audios/a.wav')
```

```python
gerarGraficoMFCC('audios/ai.wav')
```
