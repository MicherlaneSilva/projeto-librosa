```python
import numpy as np 
import matplotlib.pyplot as plt
import librosa
import librosa.display
import scipy


sound_file = 'audios/audiotemistocles.wav'
y, sr = librosa.load(sound_file, sr = 44100, duration = 0.020)

a = librosa.lpc(y, 2)
b = np.hstack([[0], -1 * a[1:]])
y_hat = scipy.signal.lfilter(b, [1], y)

fig, ax = plt.subplots()
ax.plot(y)
ax.plot(y_hat, linestyle='--')
ax.legend(['y', 'y_hat'])
ax.set_title('LP Model Forward Prediction')
```
